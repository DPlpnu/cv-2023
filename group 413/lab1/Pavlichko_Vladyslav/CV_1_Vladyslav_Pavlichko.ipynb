```python
import cv2
import matplotlib.pyplot as plt

k = 20


def show(img, k):
    plt.imshow(img)
    plt.axis('off')
    plt.title("Left shifk with k = " + str(k))
    plt.show()


def left_shift(image, k):
    newImage = image

    for x in range(0, len(newImage)):
        for y in range(len(newImage[1])):
            pixel = newImage[x][y]
            for c in range(3):
                if ((pixel[c] + k) < 256):
                    pixel[c] = pixel[c] + k
                else:
                    pixel[c] = 255

    return newImage
```

```python
def show_gray(img):
    plt.imshow(img, cmap="gray")
    plt.axis('off')
    plt.show()
```

```python
import numpy as np

img1 = cv2.imread('figures/al1.jpg', 0)
m, n = img1.shape
```

```python
def set_kernel(input):
    k = input
    kernel = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    return k, kernel, constant
```

```python
def erode(img1, kernel, constant):
    imgErode = np.zeros((m, n), dtype=np.uint8)
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = img1[i - constant:i + constant + 1, j - constant:j + constant + 1]
            product = temp * kernel

            imgErode[i, j] = np.min(product)
    return imgErode

k, kernel, constant = set_kernel(5)

imgErode = erode(img1,kernel, constant)

show_gray(img1)
show_gray(imgErode)
```

```python
k, kernel, constant = set_kernel(9)
imgErode = erode(img1,kernel, constant)
show_gray(imgErode)
```

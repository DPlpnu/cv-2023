```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the images
img = np.array(Image.open("detailed.jpg").convert('L'))

# Define the histogram shift amount
shift = 50

# Shift the histograms to the right
img_shifted = np.clip(img + shift, 0, 255)

# Plot the original and shifted images
fig, axs = plt.subplots(1, 2, figsize=(8,8))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Image 1')
axs[1].imshow(img_shifted, cmap='gray')
axs[1].set_title('Image 1 Shifted')
plt.show()
```

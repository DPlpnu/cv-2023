```python

import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(img, gamma):
    img_normalization = img / 255.0
    img_corrected = np.power(img_normalization, gamma)
    img_corrected = img_corrected * 255
    img_corrected = img_corrected.astype(np.uint8)
    return img_corrected

def visualisation(img,gamma):
    img = plt.imread(img)
    img1 = gamma_correction(img,gamma)
    fig, axs = plt.subplots(2, figsize=(10, 10))
    axs[0].imshow(img)
    axs[0].set_title('Original')
    axs[1].imshow(img1)
    axs[1].set_title('Gamma-corrected')
    plt.show()
```

#Зображення з різним контрастом

```python
visualisation("Image/sobaka.jpg",1.5)
```

```python
visualisation("Image/sobaka2.jpg",1.5)
```

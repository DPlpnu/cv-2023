```python
import numpy as np
from PIL import Image
```

```python
img = Image.open('Image.jpg')
new_image = img.resize((500, 500))
img_arr = np.array(new_image)
```

```python
new_image
```

```python
def median_filter_pixel(img, i, j, kernel_size):    
    kernel = img[max(0, i-kernel_size):min(img.shape[0], i+kernel_size+1), 
                 max(0, j-kernel_size):min(img.shape[1], j+kernel_size+1)]    
    median = np.median(kernel)
    print (kernel)
    return median
```

```python
def median_filter(img, kernel_size):    
    filtered_img = np.zeros_like(img)   
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):            
            filtered_img[i, j] = median_filter_pixel(img, i, j, kernel_size)    
    return filtered_img
```

```python
filtered_img_3 = median_filter(img_arr, 3)
filtered_img_5 = median_filter(img_arr, 5)
filtered_img_7 = median_filter(img_arr, 7)
```

```python
image1=Image.fromarray(filtered_img_3).save('image_filtered_3.jpg')
image2=Image.fromarray(filtered_img_5).save('image_filtered_5.jpg')
image3=Image.fromarray(filtered_img_7).save('image_filtered_7.jpg')
```

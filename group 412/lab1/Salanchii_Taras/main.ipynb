```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def increaseContrast(img, s, a, t=3):
    height, width, channels = img.shape

    out = np.zeros((height, width, channels), dtype=np.uint8)

    for y in range(height):
        for x in range(width):
            for c in range(channels):
                new_value = int(a * (img[y, x, c] + s) - t)

                if new_value < 0:
                    new_value = 0
                elif new_value > 255:
                    new_value = 255

                out[y, x, c] = new_value

    return out.astype(np.uint8)
```

```python
def makeExperiment(imgPath,s,a,t):
  img = cv2.imread(imgPath)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  result =  increaseContrast(img, s, a,t)


  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
  ax1.imshow(img)
  ax1.set_title('Original Image')
  ax2.imshow(result)
  ax2.set_title('Edited Image')

  ax1.axis('off')
  ax2.axis('off')


  plt.show()
```

```python
makeExperiment('img_1.jpg', 40, 1.2, 0)
makeExperiment('img_1.jpg', 20, 1.6, 0)
```

```python
makeExperiment('img_2.jpg', 40, 1.2, 0)
makeExperiment('img_2.jpg', 20, 1.6, 0)
```

```python
makeExperiment('img_3.jpg', 20, 1.2, 0)
makeExperiment('img_3.jpg', 0, 1.4, 0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma(img, gamma=1.0):
    img_norm = img / 255.0
    img_gamma = np.power(img_norm, gamma)
    img_corrected = img_gamma * 255.0
    img_corrected = img_corrected.astype(np.uint8) # to save max 8 bit in each cell
    return img_corrected
```

```python
def plot_comparison(original, new):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    ax[0].imshow(original)
    ax[0].set_title('Image 1')
    ax[0].axis('off')
    ax[1].imshow(new)
    ax[1].set_title('New Image')
    ax[1].axis('off')
    plt.show()
```

```python
img_path = 'drive/MyDrive/cv/lab1/cv1_1.jpg'
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_corrected_1 = gamma(img, gamma=0.4)
plot_comparison(img, img_corrected_1)
```

```python
img_corrected_2 = gamma(img, gamma=1.8)
plot_comparison(img, img_corrected_2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def adjust_contrast(image_path, a, c=128, b=0):

    image = Image.open(image_path)

    image = image.convert('L')

    im_array = np.array(image)

    # brightness compression function
    def adjust(x):
       return a * (x - c) + c + b

    new_values = np.vectorize(adjust)(im_array) #.astype(np.uint8)


    new_image = Image.fromarray(new_values)

    # new_image.save('out_image.jpg')

    return new_image

images = ['gradient.jpeg', 'shrek_image.png']

for image in images:

  fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12, 6))
  ax[0].imshow(np.array(Image.open(image).convert('L')), cmap='gray')
  ax[0].set_title('Original')
  ax[0].axis('off')
  
  ax[1].imshow(adjust_contrast(image, 0.8), cmap='gray')
  ax[1].set_title('Lowered Contrast')
  ax[1].axis('off')
    
  ax[2].imshow(adjust_contrast(image, 0.5), cmap='gray')
  ax[2].set_title('Even Lower Contrast')
  ax[2].axis('off')

    
  ax[3].imshow(adjust_contrast(image, 0.5, 255), cmap='gray')
  ax[3].set_title('Center of Adjustment at 255')
  ax[3].axis('off')


plt.show()
```

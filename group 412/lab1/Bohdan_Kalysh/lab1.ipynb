```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
```

```python
def showImage(image):
  plt.figure(figsize=(20, 10))
  plt.imshow(image)
```

```python
man_image = plt.imread("photos/man.png")
car_image = plt.imread("photos/car.png")
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')
plt.imshow(np.hstack([man_image, car_image]))
```

```python
def set_gamma(image: np.ndarray, gamma: float):
  return np.power(image, 1/gamma)
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')
plt.imshow(np.hstack([
    set_gamma(man_image, 0.8), 
    set_gamma(car_image, 0.8)
    ]))
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')
plt.imshow(np.hstack([
    set_gamma(man_image, 0.5), 
    set_gamma(car_image, 0.5)
    ]))
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')
plt.imshow(np.hstack([
    set_gamma(man_image, 0.2), 
    set_gamma(car_image, 0.2)
    ]))
```
